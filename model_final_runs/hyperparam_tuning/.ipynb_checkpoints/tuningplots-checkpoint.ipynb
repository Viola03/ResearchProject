```python
# Baselinemodel

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

plt.rcParams["font.family"] = "Times New Roman"
plt.rcParams["axes.unicode_minus"] = False  # Use hyphen (-) instead of the missing Unicode minus sign
```

```python
def moving_average(data, window_size=50):
    return np.convolve(data, np.ones(window_size)/window_size, mode='valid')
```

```python
base = pd.read_csv('/users/zw21147/ResearchProject/model_final_runs/hyperparam_tuning/losses_data.csv')
bs75 = pd.read_csv('/users/zw21147/ResearchProject/model_final_runs/hyperparam_tuning/bs75.csv')
bs5 = pd.read_csv('/users/zw21147/ResearchProject/model_final_runs/hyperparam_tuning/bs5.csv')
bs1000 = pd.read_csv('/users/zw21147/ResearchProject/model_final_runs/hyperparam_tuning/bs1000.csv')
bs100 = pd.read_csv('/users/zw21147/ResearchProject/model_final_runs/hyperparam_tuning/bs100.csv')
```

```python
base.columns
```

```python
fig = plt.subplot()

window_size = 50

plt.plot(base['iteration'], base['kl_loss'], alpha=1, label='KL Loss', c='tab:blue')
plt.plot(base['iteration'], base['reco_loss'], alpha=1, label='Reco Loss', c='tab:orange')

# plt.plot(base['iteration'][:-(window_size-1)], moving_average(base['kl_loss'], window_size), alpha=1.0, c='tab:blue')
# plt.plot(base['iteration'][:-(window_size-1)], moving_average(base['reco_loss'], window_size), alpha=1.0, c='tab:orange')


plt.xlabel('Iterations')
plt.ylabel('Loss Value')
plt.title('Losses')
plt.ylim(0,150)
plt.grid(alpha=0.5)
plt.legend()
```

```python
fig = plt.subplot()
plt.plot(bs5['iteration'], bs5['total_loss'], alpha=0.2, label='batch_size = 5', c='tab:blue')
plt.plot(bs75['iteration'][:15000], bs75['total_loss'][:15000], alpha=0.2, label='batch_size = 75', c='tab:orange')
plt.plot(bs100['iteration'][:15000], bs100['total_loss'][:15000], alpha=0.2, label='batch_size = 100', c='tab:pink')

plt.plot(base['iteration'], base['total_loss'], alpha=0.2, label='batch_size = 256', c='tab:green')
plt.plot(bs1000['iteration'], bs1000['total_loss'], alpha=0.2, label='batch_size = 1000', c='tab:purple')

# Smooth the loss functions
window_size = 100  # Adjust window size for desired smoothness
plt.plot(bs5['iteration'][:-(window_size-1)], moving_average(bs5['total_loss'], window_size), alpha=1.0, c='tab:blue')
plt.plot(base['iteration'][:-(window_size-1)], moving_average(base['total_loss'], window_size), alpha=1.0, c='tab:green')
plt.plot(bs75['iteration'][:15000][:-(window_size-1)], moving_average(bs75['total_loss'][:15000], window_size), alpha=1.0, c='tab:orange')
plt.plot(bs100['iteration'][:15000][:-(window_size-1)], moving_average(bs100['total_loss'][:15000], window_size), alpha=1.0, c='tab:pink')
plt.plot(bs1000['iteration'][:-(window_size-1)], moving_average(bs1000['total_loss'], window_size), alpha=1.0, c='tab:purple')

plt.xlabel('Iterations')
plt.ylabel('Total Loss')
plt.title('Varying Batch Size')
plt.ylim(0,200)
plt.grid(alpha=0.5)
plt.legend()
```
